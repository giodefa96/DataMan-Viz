Come per il notebook delle serie-tv abbiamo applicato gli stessi procedimenti.
Quindi in questo caso abbiamo realizzato un ciclo for in modo da realizzare una DF prova che contenesse tutte le informazioni provenienti dall'API di Monkey Learn

```python
prova = pd.DataFrame()
for i in range(0,len(film)):
    data = []
    data = [film['Critica'][i]]
    model_id = 'cl_pi3C7JiL'
    result = ml.classifiers.classify(model_id, data)
    prova = prova.append(result.body)
```

```python
prova
```

Abbiamo realizzato le nostre due Funzioni per estrapolare le informazioni di cui necessitavamo

```python


def metti(x):
    re_comp1 = re.compile("(Negative)")
    re_comp2 = re.compile('(Positive)')
    m = re_comp1.search(x)
    m_a = re_comp2.search(x)
    
    if m:
        return str(m.group(1))
    if m_a:
        return str(m_a.group(1))
    
```

```python
def conf(x):
    re_comp1 = re.compile('(\d?\.\d+)')
    m = re_comp1.search(x)
    
    if m:
        return float(m.group(1))
```

Abbiamo applicato le nostre funzioni

```python
prova['confidence'] = prova['classifications'].apply(conf)
```

```python
prova['tag_name'] = prova['classifications'].apply(metti)
```

```python
prova['classifications'] = prova['classifications'].astype(str)
```
